```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid", color_codes=True)
from statsmodels.graphics.mosaicplot import mosaic
from collections import Counter
```

```python
df = pd.read_csv("../dataset_diabetes/diabetic_data.csv")
```

```python
df.head(10)
```

```python
set(df['age'])
```

```python
#outcome variables: readmitted to the hospital >30, <30, or no
set(df['readmitted'])
```

```python
sns.boxplot(x="readmitted", y="time_in_hospital", hue="gender", data=df)
```

```python
sns.boxplot(x="readmitted", y="time_in_hospital", hue="race", data=df)
```

```python
mosaic(df, ['readmitted','gender'])
```

```python
set(df['race'])
```

```python
sample_size = range(len(df))
sns.boxplot(x="readmitted", y=sample_size, hue="race", data=df)
```

```python
sample_size  
```

```python
sns.factorplot(x='readmitted', y=sample_size, hue="race", data=df)
```

```python
Counter(df['race'])
```

```python
labels, values = zip(*Counter(df['race']).items())
indexes = np.arange(len(labels))
width = 1
plt.bar(indexes, values, width)
plt.xticks(indexes + width * 0.5, labels)
plt.show()
```
